```python
import numpy as np
import rasterio
import matplotlib.pyplot as plt

# Set a flooding threshold (elevation in meters)
flood_threshold = 0.2 # Adjust based on your data and scenario

# Open the DEM file
with rasterio.open("USGS_1_n28w083_20221019.tif") as src:
    dem_data = src.read(1)
    
    # Mask out no-data values, if present (like -999999.0)
    dem_data = np.ma.masked_equal(dem_data, -999999.0)

    # Create a flood mask (True where elevation is below threshold)
    flood_mask = dem_data < flood_threshold

# Plot the DEM with a colormap for elevation
plt.imshow(dem_data, cmap="terrain")
plt.colorbar(label="Elevation (m)")
plt.title("DEM with Flooding Overlay")

# Overlay the flood mask (in blue color)
plt.imshow(flood_mask, cmap="Blues", alpha=0.5)  # Adjust alpha for transparency

# Display the plot
plt.show()
```
